# Gradient descent 방식의 학습

```python

x = [0,1,2,3]
y = [3,2,8,10]
# y=func3(x,w)의 형태로 w를 예측하는것이 목표

def func3(x,w):
 return w[0]+w[1]*x+w[2]*x**2+w[3]*x**3

def gradient_descent(x, y, w_init, lr, iters):
  w = w_init
  for _ in range(iters):
    y_hat = func3(x, w)
    grad = grad_mse(y, y_hat, w)
    w = w - lr * grad # 리스트들이 끼리끼리 한번에 덧셈연산 처리가 된다
    print('w:', w)
    # print('grad:',grad)
    # print loss
    loss = np.sum((y_hat-y)**2)
    print('error:', loss)

  return w

# mean squared error 함수
# 사실 밑의 코드는 1/2로 나눠서 SSE다. 크게 상관은 없음
def grad_mse(y, y_hat, w):

  grad = np.zeros_like(w)
  # 각각의 식에 대하여 직접 미분해 보기
  grad[0] = np.sum((y_hat-y))
  grad[1] = np.sum((y_hat-y) * x)
  grad[2] = np.sum((y_hat-y) * x**2)
  grad[3] = np.sum((y_hat-y) * x**3)
  return grad


x = np.array([0, 1, 2, 3])
y = np.array([3, 2, 8, 10])
w_init = np.array([-2,-1,0.5,0.5])
lr = 1e-3
iters = int(1e5)

w = gradient_descent(x, y, w_init, lr, iters)

print("Learned weights:", w)
```
